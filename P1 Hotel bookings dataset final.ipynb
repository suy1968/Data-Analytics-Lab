```python
import pandas as pd
import numpy as np
```

#### READING THE DATASET AND PRINTING RANDOM 10 ROWS

```python
data=pd.read_csv("hotel_bookings.csv")
data.sample(10)
```

```python
df=pd.DataFrame(data)
df.head()
```

#### CHECKING FOR NULL VALUES PRINTING SUM OF NULL VALUES IN EACH COLUMN

```python
df.isnull().sum()
```

```python
print("Total null values = ",df.isnull().sum().sum())
```

#### REPLACEING NAN WITH GIVEN VALUES

```python
df['children'] = df['children'].fillna(0)
df['children'].isnull().sum()
```

```python
df['country'] = df['country'].fillna( method='ffill')
df['country'].isnull().sum()
```

```python
df['agent'] = df['agent'].replace(np.nan,df['agent'].mean())
df['country'].isnull().sum()
```

#### SHOWING SUM OF NULL VALUES IN ALL COLUMNS AFTER REMOVING NULL VALUES

```python
df.isnull().sum()
```
